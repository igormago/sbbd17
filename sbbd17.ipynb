```python
#Importando algumas bibliotecas importantes
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando alguns classificadores 
from sklearn.naive_bayes import GaussianNB 
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier

# import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
# warnings.filterwarnings("ignore")
```

```python
#Lendo Dataframe de um arquivo CSV
df = pd.read_csv('df_performances.csv')
```

```python
#Quais colunas tem esse dataframe?
df.columns
```

```python
#Olhando as primeiras linhas dos dados
df
```

```python
#Inicializando um classificador 
#Mais opÃ§Ãµes em: http://scikit-learn.org/stable/

clf = DecisionTreeClassifier()
clf = svm.SVC()
clf = GaussianNB()
clf = MLPClassifier()
clf = KNeighborsClassifier()
clf = RandomForestClassifier()
clf = GradientBoostingClassifier()
```

```python
#df = df.fillna(0)

#Dividindo os Dados em Treino e Teste
train = df[(df.s_year < 2016)]
test = df[(df.s_year == 2016)]

#Selecionando algumas features para prediÃ§Ã£o
features = ['h_num_wins','a_num_wins']

#Selecionando a variÃ¡vel alvo
target = 'm_column_result'

#Dividindo o conjunto de treinamento (features e alvo)
train_x = train[features]
train_y = train[target]

#Dividindo o conjunto de teste (features e alvo)
test_x = test[features]
test_y = test[target]

#Treinando o classificador
clf.fit(train_x,train_y)

#Verificando a acurÃ¡cia do modelo treinado com o conjunto de teste
score = clf.score(test_x, test_y)

print(score)
```

```python
hits = test[test[target] == test['predict']]
errors = test[test[target] != test['predict']]

print(len(hits))
print(len(errors))

print(hits['m_odd_home'].sum())
```
